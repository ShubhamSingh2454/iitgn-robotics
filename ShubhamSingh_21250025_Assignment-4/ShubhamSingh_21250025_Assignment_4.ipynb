```python
import numpy as np
import math 

#Shubham Singh
#21250025
#Assignment-4
#Task-1
 
# Define the desired position of the end effector

Px = float(input("End effector position in X direction:"))
Py = float(input("End effector position in y direction:"))
Pz = float(input("End effector position in z direction:"))

d1_ = int(input("Enter the value of d1:"))
theta1 = math.atan2(Px, Py)

r = math.sqrt(Px*Px + Py*Py)
s = Pz- d1_
theta2 = math.atan2(r, s)

d3 = math.sqrt((Px*Px +Py*Py)+((Pz - d1_)*(Pz - d1_)))
angle1 = math.degrees(theta1)
angle2 = math.degrees(theta2)

theta12 = math.pi + math.atan2(Px, Py)
theta22 = math.pi - math.atan2(r, s)

angle3 = math.degrees(theta12)
angle4 = math.degrees(theta22)

if d1_ == 0:
    print("Value of Theta_1 in degrees ", angle1, "\n")
    print("Value of Theta_2 in degrees",angle2, "\n")

#If there is an offset d1 not equal to zero, then the wrist centre cannot intersect the Zo axis.
#In this case there will be in general two solutions for theta1 & theta2.
elif d1_ >= 1:

    print("Value of Theta_1 in degrees:", angle1 ,"\n")
    print("Value of Theta_2 in degrees:",angle2, "\n")
    print("Value of Second solution of Theta_1 in degrees:", angle3, "\n")
    print("Value of Second solution of Theta_2 in degrees:",angle4, "\n")
print("The value of linear distance d3:", d3)
```

```python
#Shubham Singh
#21250025
#Assignment-4
#Task-3


import numpy as np
import math

n=int(input("Number of links: "))
dh=[0]*n
R_or_P=[0]*n

print("Link,  theta_i,  d_i,  a_i ,  alpha_i")
for i in range(n):
    dh[i]=[eval(j) for j in input(str(i+1)+" "*6).split()]
lst=[int(i) for i in input("Prismatic joints if any (leave empty if none): ").split()]
for i in lst:
    R_or_P[i-1]=1

A=[0]*n
for i in range(n):
    theta, d, a, alpha = dh[i]
    A[i]=np.array([[np.cos(theta), -np.sin(theta)*np.cos(alpha), np.sin(theta)*np.sin(alpha), a*np.cos(theta)],
                   [np.sin(theta), np.cos(theta)*np.cos(alpha), -np.cos(theta)*np.sin(alpha), a*np.sin(theta)],
                   [0, np.sin(alpha), np.cos(alpha), d],
                   [0, 0, 0, 1]])
ans=np.identity(4)
Z=[0]*(n+1)
Z[0]=np.array([[0],
               [0],
               [1]])

O=[0]*(n+1)
O[0]=np.array([0, 0, 0])

for i in range(n):
    ans=ans@A[i]
    Z[i+1]=ans[0:3, 0:3]@Z[0]
    O[i+1]=ans[0:3,3].T
J=[0]*n
for i in range(n):
    if R_or_P[i]==1:
        J[i]=np.concatenate((Z[i],np.array([[0,0,0]]).T),axis=0)
    else:
        cross=np.cross(Z[i].T,(O[n]-O[i]).T)
        J[i]=np.concatenate((cross.T,Z[i]),axis=0)

final_J=J[0]
for i in range(1,n):
    final_J=np.concatenate((final_J,J[i]),axis=1)

print("Jacobian Matrix is:", final_J)

dots=input("Enter x_dot: ("+str(n)+" values) ")
print()
x_dots=[int(i) for i in dots.split()]          

print("Joint Velocities:")
if n == 6:
    ki = np.linalg.inv(final_J)

    q_dot = ki@x_dots
    print(q_dot)

else:
    print(" Jacobian is not invertible")
```
